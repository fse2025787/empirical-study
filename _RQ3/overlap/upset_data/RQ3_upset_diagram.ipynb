## REN

```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import itertools
from upsetplot import generate_counts, plot


df = pd.read_csv('../RQ2_tools_five_swc_id.csv')


SWCIDs = ["SWC-107", "SWC-106", "SWC-115", "SWC-105", "SWC-101"]

for i in SWCIDs:
    if i != "SWC-107":continue

    filtered_df = df[df['SWC_ID'] == i]


    tools = filtered_df['tools'].unique()
    

    categories = list(itertools.product([False, True], repeat=tools.size))

    filtered_categories = [comb for comb in categories if not (sum(comb) in [0])]

    sorted_categories = sorted(filtered_categories, key=lambda x: sum(x))

    tools_format = {
        'ethainter':'Ethainter',
        'madmax':'MadMax',
        'mythril': 'Mythril',
        'semgrep': 'Semgrep',
        'slither': 'Slither',
        'smartcheck':'SmartCheck',
        'solhint': 'Solhint',
        'soliditycheck':"SolidityCheck",
        'vandal': 'Vandal',
        'verismart':'VeriSmart'
    }

    formatted_tools = [tools_format[tool] for tool in tools]
    

    index = pd.MultiIndex.from_tuples(sorted_categories, names=formatted_tools)

    df = pd.read_csv(f'./upset_data/RQ2__upset_data_{i}.csv')


    data_dict = df.set_index('Intersection Name')['Intersection Count'].to_dict()
    

    def map_to_index(tool_combination, tools):
        boolean_combination = [tool in tool_combination for tool in tools]
        return tuple(boolean_combination)


    mapped_values = {map_to_index(k, tools): v for k, v in data_dict.items()}


    s = pd.Series(mapped_values, index=index, name="value")

    matplotlib.rcParams["font.size"] = 8

    plot(s, show_counts="{:,}",show_percentages="{:.0%}")

    plt.savefig(f'./RQ3_upset_graph_{i}.png', dpi=600, bbox_inches='tight')
```
