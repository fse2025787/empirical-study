```python
import pandas as pd
import matplotlib.pyplot as plt

SWCIDs = ["SWC-107", "SWC-106", "SWC-115", "SWC-105", "SWC-101"]

for i_swc in SWCIDs:
    
    df = pd.read_csv(f'./Jaccard_Similarity_results_{i_swc}.csv')

    columns_to_plot = df.columns[2:]
    num_tools = df['num_tools']
    plt.figure(figsize=(5, 3))

    for index, row in df.iterrows():
        plt.plot(columns_to_plot, row[columns_to_plot], marker='o', label=f'Line {index + 2}')

    plt.title(f'Jaccard Similarity results {i_swc}')
    plt.xlabel('num_tools')
    plt.ylabel('Average')
    plt.xticks(num_tools)
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

SWCIDs = ["SWC-107", "SWC-106", "SWC-115", "SWC-105", "SWC-101"]
label_ = {"SWC-107":"REN", "SWC-106":"USI", "SWC-115":"TXO", "SWC-105":"UEW", "SWC-101":"IOU"}

plt.figure(figsize=(5, 3))  

for i_swc in SWCIDs:
    df = pd.read_csv(f'./Jaccard_Similarity_results_{i_swc}.csv')

    filtered_df = df[df['num_tools'] == 2]

    columns_to_plot = filtered_df.columns[2:]
    
    for index, row in filtered_df.iterrows():
        plt.plot(columns_to_plot, row[columns_to_plot], marker='o', label=f'{label_[i_swc]}')

plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.0), ncol=3,fontsize=10)
new_labels = ['0.4.x', '0.5.x', '0.6.x', '0.7.x', '0.8.x']  
plt.xticks(ticks=range(len(new_labels)), labels=new_labels,fontsize=16)
plt.yticks(fontsize=16)

plt.tight_layout()

plt.savefig('./RQ3_J2.png', dpi=600, bbox_inches='tight')
```
